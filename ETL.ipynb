### Hora

```python
df_homicidios['Hora']=pd.to_datetime(df_homicidios['Hora'], format='%H:%M:%S').dt.time
```

Se crea una función para transformar las columnas Latitud y Longitud en float y en el caso de no poder reemplaza el valor por NaN

```python
valores=list()
def transformar_float(x):
    try:
        return float(x)
    except:
        valores.append(x)
```

```python
df_homicidios['Latitud'] = df_homicidios['Latitud'].apply(transformar_float)
```

```python
valores
```

```python
valores=list()
def transformar_float(x):
    try:
        return float(x)
    except:
        valores.append(x)
```

```python
df_homicidios['Longitud'] = df_homicidios['Longitud'].apply(transformar_float)
```

```python
valores
```

Como se aprecia los valores que no se puediero convertir en *float* son puntos, por lo que se remplazaran por **None**

```python
def reemplazar(x):
    try:
        return float(x)
    except:
        return None
```

```python
df_homicidios['Latitud'] = df_homicidios['Latitud'].apply(reemplazar)
df_homicidios['Longitud'] = df_homicidios['Longitud'].apply(reemplazar)
```
